```python
import easyocr
import pyttsx3
```

```python
reader = easyocr.Reader(["en"])
```

```python
result = reader.readtext("assets/testtext.png")
```

```python
text = " ".join([res[1] for res in result])
```

```python
print("Text: ", text)
```

```python
engine = pyttsx3.init()
engine.say(text)
engine.runAndWait()
```
