```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("penguins") #df = dataframe
```

```python
df
```

```python
sns.pairplot(data = df, hue = "species")
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df
```

```python
df = df.drop(["island", "sex"], axis = 1) #eckige Klammern fÃ¼r Spalten lÃ¶schen
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(["species"], axis = 1)
```

```python
X
```

```python
y = df["species"]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 119)
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier(n_neighbors = 10) #clf = classifier
```

```python
clf.fit(X_train, y_train) #fit = trainieren
```

```python
predictions = clf.predict(X_test) #predict = vorhersage
```

```python
predictions
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(predictions, y_test) #80% genauigkeit
```

```python
# Using k=10, the KNN model reached 77% accuracy (test size = 0.2, random state = 119)
```
